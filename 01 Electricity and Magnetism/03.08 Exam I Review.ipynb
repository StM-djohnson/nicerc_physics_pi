#### Vocabulary

**Microcontroller** – A small computer on a single integrated circuit that can be programmed to read sensors, make decisions, and orchestrate devices that control moving parts.

**Flowchart** – Graphical representation of a computer program.

**Schematic Diagram** – Graphical representation of an electronic circuit.

**LED** – Light Emitting Diode

**Load** – Something that uses energy (in a circuit).

**Source** – Something that supplies energy (in a circuit).

**Charge** – A measure of electrostatic force; can be either positive (+) or negative (‐).

**Circuit** – A closed path that allows charge to flow.

**Electricity** – The movement of charge (usually electrons).

**Conductor** – Material through which electrons travel easily. They contain a large number of free electrons.

**Insulator** – Material through which electrons travel poorly or not at all.

**Semiconductor** – Material that can act like a conductor or an insulator depending on certain conditions.

**Voltage (V)** – An electromotive force (emf), analogous to pressure, that drives the flow of charge. The work done in moving a charge from point A to point B in an electrical circuit, measured in joules (J), divided by the charge, measured in coulombs (C):
$$ 1V = \frac{1J}{1C} $$

**Joule** ‐ A unit of energy (SI system) equal to:
$$ \frac{1kg \cdot m^2}{s^2} $$

**Coulomb** – A quantity of charge equal to 6.24 x 1018 electrons.

**Current (i)** – The quantity of electrons passing a given point in a given time. Measured in coulombs per second. 

$$ \frac{1 coulomb}{1 second} = 1 amperes (A)$$

**Resistance (r)** – Frictional losses that occur when electrons flow through a material. The property of a material that describes how easily charge may pass through it. 

$$1\Omega = \frac{1V}{1A}$$

**Photoresistor** – A resistor whose resistance varies with exposure to light.

**Potentiometer** – Variable resistor that effectively creates two resistors with a center “tap” (wiper), such that the values of the two resistors separated by the wiper always add up to the value of the potentiometer.

**(Resistor) Tolerance** – The amount (percentage) of possible variation from the specified value.

**Analog Signal (Continuous)** – Signals (e.g., voltage, current) that are continuous in both time and value (magnitude).
