# Electricity and Magnetism 
## 3 Electricity & Circuitry
### 03 Statistics II

Class Activity:
* Each student will begin with 4 – 220 Ω resistors.

* With the multimeters, measure the resistance across the resistors. Make sure that the multimeter dial is set to measure resistance (Ω).

* Record the resistances measured on a data sheet which will be used to create a class data set.

* Gather all data from each classmate, and record the values in an Google Sheets spreadsheet.

---

![resist1](../images/03.03.resist.01.png)

---

* Perform statistical analyses to understand the distribution of the resistance values. Choose a cell below the raw data and use the =AVERAGE function to calculate the mean/average resistance value.

---

![resist2](../images/03.03.resist.02.png)

---

* Choose another cell and use the =STDEV function to calculate the standard deviation of the data set. **Note:** The STDEV function is the Google Sheets function used to calculate the sample standard deviation. If you have more than 30 samples, then you would use the population standard deviation function, which is STDEVP.

#### Class Discussion:

What have you learned about resistors from these statistical analyses?

1. Is the mean of the resistors around 220 Ω? Explain.

2. Is the standard deviation large or small? What does this value say about the distribution of the resistance values?

3. Can you relate the standard deviation and mean with the histogram? Explain.

4. Knowing the tolerance of the resistor (as indicated on the color code), did the resistance values stay within the tolerance allotted for the resistors?
