# Honors Physics 2018/19

## Course Outline 

### INTRODUCTION 
#### Unit 1  Electricity and Magnetism (13 weeks) 
1. Introduction to Physics and Raspberry Pi Programming
2. Electric Forces, fields and potential (Coulomb’s law, electric force, fields and potential)
3. Basic Circuits (DC circuits, steady state RC circuits, Ohm’s Law and Kirchhoff’s Law)
  1. Series Circuits
  2. Parallel Circuits
  3. Resistance
4. Magnetic Fields and Magnetic Forces and Magnetic Induction
5. Electromagnetism (flux, forces on moving charges, current carrying wires and their magnetic fields)
6. Motors

#### Unit 2: Work and Mechanics (5.5 weeks) 
1. Falling Objects
2. 1D Particle Motion
3. 2D Particle Motion
4. Newton’s Laws of Motion
5. Force 
6. Static Equilibrium
7. Machines
8. Energy
  1. Work and Power
  2. Conservation of Energy
  3. Efficiency
  4. Spring and Elastic Energy
  5. Kinetic Energy
9. Vector Stress (truss) Analysis

#### Unit 3: Light and Optics (4 weeks) 
1. Introduction (Electro magnetic spectrum, Reflection, refraction, diffraction)
2. Polarization and Optics (geometric and physical)
3. Mirrors and Lenses (Images formed by mirrors and lenses, ray diagrams)
4. Waves Fronts

#### Unit 4: Waves and Sound (3 weeks) 
1. Properties of Waves
2. Waves at Boundaries
3. Sound Waves
4. Doppler Shift
5. Physics of Music

#### Unit 5: Thermal Fluids and Thermodynamics (7 weeks) 
1. Density and pressure and Pascal’s principle
2. Fluid Dynamics Buoyancy – Archimedes’ Principle
3. Fluid flow continuity
4. Bernoulli's equation
5. Temperature and Heat
6. Thermodynamics (conduction, convection and radiation)
7. Laws of Thermodynamics (entropy, ideal gases and kinetic theory)

#### Unit 6: Modern Physics (4 weeks) 
1. Atomic Physics and Quantum Effects
  1. Photons and the Photoelectric effect 
  2. X-ray production
  3. Electron energy levels 
  4. Compton scattering
  5. Wave nature of matter 
2. Nuclear Physics
  1. Atomic mass, mass number, atomic number 
  2. Mass defect and nuclear binding energy
  3. Nuclear processes 
  4. Mass-Energy Equivalence and Conservation of Mass and Energy
